```r
datos <- data.frame(y = c(
    rep(0,1),
	rep(1,2),
	rep(2,3),
	rep(3,4),
	rep(4,13),
	rep(5,12),
	rep(6,11),
	rep(7,10),
	rep(8,9),
	rep(9,10),
	rep(10,11),
	rep(11,12),
	rep(12,13),
	rep(13,4),
	rep(14,3),
	rep(15,2),
	rep(16,1)
),
                   x = c(
    rep(0,2),
    rep(1,6),
    rep(2,9),
    rep(3,9),
    rep(4,9),
    rep(5,13),
    rep(6,17),
    rep(7,15),
    rep(8,11),
    rep(9,9),
    rep(10,9),
    rep(11,8),
    rep(12,4)
))

ggplot(datos, aes(x = x)) + 
    geom_histogram(aes(y = after_stat(density)), fill = "gainsboro", binwidth = 1) + 
    geom_density(size = 2, color = "steelblue") + theme_minimal()
ggplot(datos, aes(x = y)) + 
    geom_histogram(aes(y = after_stat(density)), fill = "gainsboro", binwidth = 1) + 
    geom_density(size = 2, color = "steelblue") + theme_minimal()
```

```r
apply(X = datos, MARGIN = 1, FUN = print)
```

```r
distancia <- function(m){
    return(apply(X = as.data.frame(datos$x), MARGIN = 1, FUN = function(y) return(abs(m-y))))
}
datos_distancias <- as.vector(apply(X = as.data.frame(datos$y), MARGIN = 1, FUN = distancia))
datos_distancias
```
